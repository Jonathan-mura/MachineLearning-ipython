```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
colors = ('r', 'g', 'b', 'k')
for c in colors:
    x = np.random.sample(30)
    y = np.random.sample(30)
    ax.scatter(x, y, 0, zdir='y', c=c)
```
